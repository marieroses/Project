```python
from sqlalchemy import create_engine
import pandas as pd

engine=create_engine('sqlite:///countries_database.sqlite')
con=engine.connect()
rs=con.execute("Select * from 'countries of the world'")
df=pd.DataFrame(rs.fetchall())
df.columns=rs.keys()
con.close()

print(df.head())
```

```python
engine=create_engine('sqlite:///countries_database.sqlite')
con=engine.connect()
rs=con.execute("Select Deathrate, GDP , Literacy  from 'countries of the world'")
df=pd.DataFrame(rs.fetchall())
df.columns=rs.keys()
con.close()

print(df.head())
```
