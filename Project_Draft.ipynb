```python
import pandas as pd


filename='C:/Users/mschm/OneDrive/Bureau/UCD/Project_Related/who_suicide_statistics.csv.zip'
WHO=pd.read_csv(filename)
WHO.head()
```

```python
WHO.info()
```

```python
WHO.isna().sum()
```

```python
WHO_clean=WHO.dropna()
print(WHO_clean)
print(WHO_clean['year'].min())
print(WHO_clean['year'].max())
```

```python
WHO_by_country=WHO_clean.groupby('country')['suicides_no','population'].sum()
print(WHO_by_country.head())
```

```python
filename_bis='C:/Users/mschm/OneDrive/Bureau/UCD/Project_Related/countries.csv.'
countries=pd.read_csv(filename_bis)
countries_ind=countries.set_index('Country')
countries_ind.head()
```

```python
WHO_by_country_with_continent=WHO_by_country.merge(countries_ind,left_index=True,right_index=True,how='left')
print(WHO_by_country_with_continent.head())
```

```python
WHO_by_continent=WHO_by_country_with_continent.groupby('Region')['suicides_no','population'].sum()
print(WHO_by_continent)
```
